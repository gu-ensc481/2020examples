# Simple built-in interpolations
I'll use the interploation packages [Dierckx.jl](https://github.com/kbarbary/Dierckx.jl) that is based a very popular library.  It is standard library used in [scipy.interpolate](https://docs.scipy.org/doc/scipy/reference/interpolate.html) for example.

First, I'll make up some irreulgarly spaced data, and then show the plot.

```julia
using Pkg
Pkg.activate(".")
using Dierckx
using Plots
```

```julia
x = 4*rand(30) |> sort
# x = 0.1:0.1:4
f(x) = log(x)
y = f.(x)
;
```

```julia
# make the spline function to use later
p = Spline1D( x, y );
```

```julia
t = LinRange(0.1, 4, 50)
plot( t, f.(t), label="True curve")
scatter!(x,y, label="Data points", lw=3)
plot!(t,p.(t), label="Interpolation", linestyle=:dash, lw=3)
```
